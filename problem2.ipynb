```
import matplotlib.pyplot as plt
import numpy as np
import math

A = np.array([[3.0,0.5],[0.5,1.0]]).astype(np.float)
myu = np.array([1,2])
lamb = 2
rate = (4 - math.sqrt(5)) / 11
w = np.array([[3,-1]])
error = []

def update(w):
  y = w - rate*np.dot((A + A.T),(w - myu))
  w_new = []  
  for i in y:
    if i > rate * lamb:
      w_new.append(i - rate * lamb)
    elif i < - rate * lamb:
      w_new.append(i + rate * lamb)
    else:
      w_new.append(0)
  return np.array([w_new])

for i in range(1000):
  if i >= 1 and np.all(w[i] == w[i-1]):
    break
  w = np.append(w,update(w[i]),axis = 0)
  
x = np.arange(len(w))
y = np.array([])
for i in x:
  y = np.append(y,np.sqrt(np.power(w[i][0] - w[len(w)-1][0],2) + np.power(w[i][1] - w[len(w)-1][1],2)))

plt.plot(x,y)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.set_yscale('log')
plt.title('Problem2_1 lambda=2',fontsize= 18)
plt.xlabel('Iteration',fontsize=18)
plt.ylabel('error',fontsize=18)
print('result:{0}'.format(len(w)))
```
